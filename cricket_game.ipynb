```python
import random

cricket_condition = [6, 4, 3, 2, 1, 2, 4, 'wide', 'no ball', 'runout', 'caught', 'bowled']

player_number = int(input("   How many players do you want to play on your team?: "))

o = int(input("     Select overs (1, 2, or 3): "))
def over(overs):

  if overs == 1:
    return 6
  if overs == 2:
    return  12
  if overs == 3:
    return  18
  else:
    return "Invalid over choice"

        #exit()  # Exit the program if the over choice is



def batting(balls,players_in_team):
    runs = 0
    wickets = 0
    while balls != 0 and players_in_team != 0:  # Use 'and' instead of 'or'
        input("         >>>Click enter for bat<<<\n")
        balls -= 1
        possition = random.choice(cricket_condition)
        if isinstance(possition, int):  # Check if possition is an integer
            runs += possition
            print(f"You get {possition} runs in this ball.\n")
        if possition == 'wide' or possition == 'no ball':
            balls += 1
            runs += 1
            print(f"It's a {possition}. You get 1 extra run.\n")
        if possition == 'runout' or possition == 'caught' or possition == 'bowled':
            players_in_team -= 1
            wickets += 1
            print(f"Batsman is {possition}\n")
        print(f"Score: {runs}/{wickets}\n")
    return runs

def target():
    run = batting(over(o),player_number)
    print(f"      \n\ntarget : {run + 1}\n")
    print("----------------------------------\n\n")
    return run + 1

def computer_batting(balls,players_in_team):
    runs = 0
    wickets = 0
    while balls != 0 and players_in_team != 0:
        balls -= 1
        possition = random.choice(cricket_condition)
        if isinstance(possition, int):  # Check if possition is an integer
            runs += possition
            print(f"computer get {possition} runs in this ball.\n")
        if possition == 'wide' or possition == 'no ball':
            balls += 1
            runs += 1
            print(f"It's a {possition}. computer get 1 extra run.\n")
        if possition == 'runout' or possition == 'caught' or possition == 'bowled':
            players_in_team -= 1
            wickets += 1
            print(f"computer is {possition}\n")
        print(f"Score: {runs}/{wickets}\n")
    return runs

def com_target():
    run = computer_batting(over(o),player_number)
    print(f"      \n\ntarget : {run + 1}\n")
    print("----------------------------------\n\n")
    return run + 1

def win(target_run, chase_run):
  if chase_run >= target_run:
    return True
  if chase_run < target_run - 1:
    return False

choice = input("    choose your opponent\n enter 'a' if you want to play with computer\n enter 'b' if you want to play with another person\n ")

print("       >>> toss <<<\n")
toss = input("\nenter 't' if choose tail or 'h' for head : ")
toss_list = ['h','t']
toss_result = random.choice(toss_list)
if toss == toss_result:
  print("\n\n     you win the toss\n")
  what_you_do = input("\nenter 'x' if you want to bat first or enter 'y' if you want to run chase \n")

  if what_you_do == 'x':
    target_score = target()
    print("\n\n     opponent's turn\n")
    chase_score = batting(over(o),player_number)

    if win(target_score, chase_score):
      print("\n\n     you loss")
    elif win(target_score, chase_score) is False:
      print("\n\n     you win")
    else:
      print("\n\n   match draw")

  if what_you_do == 'y':

    if choice == 'a':
      target_score = com_target()
      print("\n\n   your turn")
      chase_score = batting(over(o),player_number)

      if win(target_score, chase_score):
        print("\n\n   you win")
      elif win(target_score, chase_score) is False:
        print("\n\n   you loss")
      else:
        print("\n\n   match draw")

    elif choice == 'b':
      target_score = target()
      print("\n\n   opponent's turn")
      chase_score = batting(over(o),player_number)

      if win(target_score, chase_score):
        print("\n\n   you win")
      elif win(target_score, chase_score) is False:
        print("\n\n   ou loss")
      else:
        print("\n\n   match draw")


elif toss != toss_result and choice == 'b':
  print("opponent win the toss")
  what_oppo_do = input("enter 'c' if opponent wants to bat first or enter 'd' if opponent wants to run chase \n")

  if what_oppo_do == 'c':
    print("\n\n   opponent wants to bat first")
    target_score = target()
    print("\n\n   your turn")
    chase_score = batting(over(o),player_number)

    if win(target_score, chase_score):
      print("\n\n   you win")
    elif win(target_score, chase_score) is False:
      print("\n\n   you loss")
    else:
      print("\n\n   match draw")


  if what_oppo_do == 'd':
    print("\n\n   opponent wants to run chase")
    target_score = target()
    print("\n\n   opponent's turn")
    chase_score = batting(over(o),player_number)

    if win(target_score, chase_score):
      print("\n\n   you loss")
    elif win(target_score, chase_score) is False:
      print("\n\n   you win")
    else:
      print("\n\n   match draw")



else:
  print("\n\n   computer win the toss")
  com_choice_list = [1,2]
  com_choice = random.choice(com_choice_list)

  if com_choice == 1:
    print("\n\n   computer wants to bat first")
    target_score = com_target()
    print("\n\n   your turn")
    chase_score = batting(over(o),player_number)

    if win(target_score, chase_score):
      print("\n\n   you win")
    elif win(target_score, chase_score) is False:
      print("\n\n   you loss")
    else:
      print("\n\n   match draw")



  else:
    print("\n\n   computer wants to run chase")
    target_score = target()
    print("\n\n   opponent's turn")
    chase_score = computer_batting(over(o),player_number)

    if win(target_score, chase_score):
      print("\n\n   you loss")
    elif win(target_score, chase_score) is False:
      print("\n\n   you win")
    else:
      print("\n\n   match draw")
```
